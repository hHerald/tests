# Кейс продуктового аналитика

## Описание

Вы стажёр-продуктовый аналитик в МобиКар (сервисе посуточной и почасовой аренды автомобилей через мобильное приложение).

**Пользователи:**

- Устанавливают приложение

- Проходят регистрацию

- Ищут авто рядом

- Бронируют и начинают поездку

**Проблема:**

Конверсия пользователей с первого входа в приложение до первой поездки упала на 15% за последние 2 месяца.

**При этом:**

- Установки выросли на 10%

- Удержание через 7 дней у зарегистрированных пользователей не изменилось

- Новых крупных обновлений не было, кроме изменений в поиске авто

- Отзывы в магазинах приложений не ухудшились
